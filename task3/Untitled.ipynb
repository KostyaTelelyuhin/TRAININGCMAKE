```python
import pygame
import sys
import random

# Инициализация Pygame
pygame.init()

# Размеры окна
window_size = (600, 400)
window = pygame.display.set_mode(window_size)
pygame.display.set_caption('Змейка')

# Цвета
white = (255, 255, 255)
black = (0, 0, 0)
red = (213, 50, 80)
green = (0, 255, 0)

# Скорость игры
clock = pygame.time.Clock()
snake_speed = 15

# Размер блока
block_size = 10

# Шрифт
font_style = pygame.font.SysFont(None, 30)

def message(msg, color):
    mesg = font_style.render(msg, True, color)
    window.blit(mesg, [window_size[0] / 6, window_size[1] / 3])

def game_loop():
    game_over = False
    game_close = False

    # Начальные координаты змейки
    x1 = window_size[0] / 2
    y1 = window_size[1] / 2

    # Изменение координат
    x1_change = 0
    y1_change = 0

    # Тело змейки
    snake_List = []
    Length_of_snake = 1

    # Координаты еды
    foodx = round(random.randrange(0, window_size[0] - block_size) / 10.0) * 10.0
    foody = round(random.randrange(0, window_size[1] - block_size) / 10.0) * 10.0

    while not game_over:

        while game_close:
            window.fill(white)
            message("Вы проиграли! Нажмите Q для выхода или C для новой игры", red)
            pygame.display.update()

            for event in pygame.event.get():
                if event.type == pygame.KEYDOWN:
                    if event.key == pygame.K_q:
                        game_over = True
                        game_close = False
                    if event.key == pygame.K_c:
                        game_loop()

        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                game_over = True
            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_LEFT:
                    x1_change = -block_size
                    y1_change = 0
                elif event.key == pygame.K_RIGHT:
                    x1_change = block_size
                    y1_change = 0
                elif event.key == pygame.K_UP:
                    y1_change = -block_size
                    x1_change = 0
                elif event.key == pygame.K_DOWN:
                    y1_change = block_size
                    x1_change = 0

        # Проверка выхода за границы
        if x1 >= window_size[0] or x1 < 0 or y1 >= window_size[1] or y1 < 0:
            game_close = True

        x1 += x1_change
        y1 += y1_change
        window.fill(black)
        pygame.draw.rect(window, green, [foodx, foody, block_size, block_size])
        snake_Head = []
        snake_Head.append(x1)
        snake_Head.append(y1)
        snake_List.append(snake_Head)
        if len(snake_List) > Length_of_snake:
            del snake_List[0]

        # Проверка столкновения с самим собой
        for x in snake_List[:-1]:
            if x == snake_Head:
                game_close = True

        # Рисование змейки
        for x in snake_List:
            pygame.draw.rect(window, white, [x[0], x[1], block_size, block_size])

        pygame.display.update()

        # Проверка, съела ли змейка еду
        if x1 == foodx and y1 == foody:
            foodx = round(random.randrange(0, window_size[0] - block_size) / 10.0) * 10.0
            foody = round(random.randrange(0, window_size[1] - block_size) / 10.0) * 10.0
            Length_of_snake += 1

        clock.tick(snake_speed)

    pygame.quit()
    sys.exit()

game_loop()
```
